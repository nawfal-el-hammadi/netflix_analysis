```python
from google.colab import drive
drive.mount('/content/drive')

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (make sure your CSV is in the 'data' folder)
df = pd.read_csv("netflix_titles.csv")

# Show first 5 rows
df.head()
# Drop rows where 'country' or 'rating' is missing
df = df.dropna(subset=['country', 'rating'])


print("Content types:", df['type'].unique())       # Movies vs TV Shows
print("Some countries:", df['country'].unique()[:20])
print("Ratings:", df['rating'].unique())


```

```python
### Data Loading and Cleaning

- Mounted Google Drive to access project files.
- Imported necessary libraries: `pandas` for data handling and `matplotlib` for visualization.
- Loaded the Netflix dataset (5k+ titles) into a pandas DataFrame.
- Dropped rows where `country` or `rating` were missing to ensure accurate analysis.
- Explored basic dataset features:
  - **Content types:** Movies vs TV Shows
  - **Sample countries:** First 20 unique countries
  - **Ratings:** Different age classifications of titles

**Insight:** The dataset is now cleaned and ready for exploration and visualization.
```

```python

#First visualizations
df['type'].value_counts().plot(kind='bar', title='Movies vs TV Shows')
plt.show()
```

```python
### Movies vs TV Shows
- This bar chart shows the distribution of content types on Netflix.
- The majority of titles are Movies, with TV Shows being fewer in comparison.
- This gives a quick overview of Netflixâ€™s content focus.
```

```python
# Count number of titles per release year
release_trend = df['release_year'].value_counts().sort_index()

# Plot trend
plt.figure(figsize=(12,6))
release_trend.plot(kind='line', marker='o')
plt.title('Netflix Titles Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.show()
```

```python
### Release Year Trend
- This plot shows how the number of Netflix titles released has changed over time.
- We can see a significant increase in titles after 2015.
- Trend indicates Netflixâ€™s rapid content expansion in recent years.
```

```python
# Group by country and type, count titles
type_country = df.groupby(['country','type']).size().unstack(fill_value=0)

# Select top 10 countries by total number of titles
top_countries = df['country'].value_counts().head(10).index
type_country_top = type_country.loc[top_countries]

# Plot stacked bar chart
type_country_top.plot(kind='bar', stacked=True, figsize=(12,6))
plt.title('Content Type by Country (Top 10 Countries)')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
### Content Type by Country (Top 10)
- This stacked bar chart shows the number of Movies vs TV Shows each of the top 10 countries produces.
- USA and India dominate both categories, while some countries focus mainly on Movies.
- Provides insight into global content distribution and Netflix production trends.
```
